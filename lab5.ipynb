```python
G = MatrixGroup([Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]]), 
                 Matrix([[1, 0, 0], [0, 1, 1], [0, 0, 1]])])


is_finite = G.is_finite()
print(f"1. Чи є група G скінченною? {'Так' if is_finite else 'Ні'}")


is_abelian = G.is_abelian()
print(f"2. Чи є група G комутативною? {'Так' if is_abelian else 'Ні'}")

elements = G.list() if is_finite else "Група нескінченна, елементи не можна перелічити"
print("3. Який вигляд мають елементи групи G?")
print(elements)


if is_finite:
    print("4. Порядки елементів групи G:")
    for elem in elements:
        print(f"Елемент:\n{elem}\nПорядок: {elem.order()}")
else:
    print("4. Група нескінченна, тому неможливо визначити порядки всіх елементів.")


is_isomorphic_to_Z2 = G.is_isomorphic(CyclicPermutationGroup(2))
is_isomorphic_to_Z4 = G.is_isomorphic(CyclicPermutationGroup(4))
print("5. Чому ізоморфна група G?")
if is_isomorphic_to_Z2:
    print("Група G ізоморфна циклічній групі порядку 2.")
elif is_isomorphic_to_Z4:
    print("Група G ізоморфна циклічній групі порядку 4.")
else:
    print("Група G не ізоморфна простим циклічним групам порядку 2 чи 4.")
```

```python
K = MatrixGroup([Matrix([[-1, 0, 0], [0, 1, 1/2], [0, 0, 1]]), 
                 Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])])

is_finite = K.is_finite()
print(f"1. Чи є група K скінченною? {'Так' if is_finite else 'Ні'}")

is_abelian = K.is_abelian()
print(f"2. Чи є група K комутативною? {'Так' if is_abelian else 'Ні'}")

generators = K.gens()
print("3. Генератори групи K:")
for gen in generators:
    print(gen)

print("4. Порядки генераторів групи K:")
for gen in generators:
    print(f"Генератор:\n{gen}\nПорядок: {gen.order()}")

is_isomorphic_to_Z2 = K.is_isomorphic(CyclicPermutationGroup(2))
is_isomorphic_to_Z4 = K.is_isomorphic(CyclicPermutationGroup(4))
print("5. Чому ізоморфна група K?")
if is_isomorphic_to_Z2:
    print("Група K ізоморфна циклічній групі порядку 2.")
elif is_isomorphic_to_Z4:
    print("Група K ізоморфна циклічній групі порядку 4.")
else:
    print("Група K не ізоморфна простим циклічним групам порядку 2 чи 4.")

print("Додаткова задача: Аналіз підгруп, утворених генераторами групи K:")
for i, gen in enumerate(generators, 1):
    cyclic_subgroup = MatrixGroup([gen])
    print(f"Підгрупа, утворена генератором {i}:\n{cyclic_subgroup}\nІзоморфізм з циклічною групою порядку {gen.order()}.")

```
