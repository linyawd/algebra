```python
D = DihedralGroup(8)
H = D.subgroup([D[2]])
print("Підгрупа H:", H)
print("H є нормальною підгрупою:", H.is_normal(D))
cosets = D.cosets(H, side='right')
print("Праві класи суміжності:")
for coset in cosets:
    print(coset)
```

```python
A5 = AlternatingGroup(5)
subgroups = A5.subgroups()

for subgroup in subgroups:
    print("Підгрупа:", subgroup)
    print("Нормальна:", subgroup.is_normal(A5))
    print("---")

print("Група A5 є простою:", A5.is_simple())
```

```python
for n in range(3, 101):
    Dn = DihedralGroup(n)
    normal_subgroups = Dn.normal_subgroups()
    orders = [H.order() for H in normal_subgroups]
    print(f"Група D_{n}, порядки нормальних підгруп: {orders}")
    print("---")


Dn_large = DihedralGroup(470448)
print("Кількість нормальних підгруп для D_470448:", len(Dn_large.normal_subgroups()))
```
