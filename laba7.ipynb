```python
#завд1
G = DihedralGroup(20)
normal_subgroups = G.normal_subgroups()

for N in normal_subgroups:
    print(f"Нормальна підгрупа порядку {N.order()}")
    
    phi = G.hom([G.gen(0), G.gen(1)], codomain=G)  
    
    kernel = phi.kernel()
    print(f"Ядро ендоморфізму (порядок {kernel.order()}): {kernel}")
    
    if kernel == N:
        print("✅ Ядро ендоморфізму збігається з нормальною підгрупою")
    else:
        print("❌ Ядро не збігається з нормальною підгрупою")

```

```python
#завд2
groups = {
    "C11": CyclicPermutationGroup(11),
    "D13": DihedralGroup(13),
    "Q8": QuaternionGroup(),
    "SL(3, Z7)": SL(3, 7)  
}

for name, G in groups.items():
    print(f"Група {name}:")
    print(f"Кількість елементів: {G.order()}")
    center = G.center()
    print(f"Центр (порядок {center.order()}): {center}")
    conjugacy_classes = G.conjugacy_classes()
    print(f"Кількість класів спряженості: {len(conjugacy_classes)}")
    representatives = [cls.representative() for cls in conjugacy_classes]
    print(f"Представники класів спряженості: {representatives}")
    print("-" * 50)

```

```python
#завд4
graphs_with_A2 = []
graphs_with_A3 = []
graphs_with_A4 = []
graphs_with_A5 = []


for g in graphs.nauty_geng("6"):  
    if g.num_verts() > 1 and g.num_edges() > 0: 
        if g.automorphism_group().is_isomorphic(AlternatingGroup(2)):
            graphs_with_A2.append(g)
        elif g.automorphism_group().is_isomorphic(AlternatingGroup(3)):
            graphs_with_A3.append(g)
        elif g.automorphism_group().is_isomorphic(AlternatingGroup(4)):
            graphs_with_A4.append(g)
        elif g.automorphism_group().is_isomorphic(AlternatingGroup(5)):
            graphs_with_A5.append(g)

print(f"Графи з автоморфізмами A2:")
if graphs_with_A2:
    for g in graphs_with_A2:
        show(g)
else:
    print("Не знайдено відповідних графів.")

print(f"Графи з автоморфізмами A3:")
if graphs_with_A3:
    for g in graphs_with_A3:
        show(g)
else:
    print("Не знайдено відповідних графів.")

print(f"Графи з автоморфізмами A4:")
if graphs_with_A4:
    for g in graphs_with_A4:
        show(g)
else:
    print("Не знайдено відповідних графів.")

print(f"Графи з автоморфізмами A5:")
if graphs_with_A5:
    for g in graphs_with_A5:
        show(g)
else:
    print("Не знайдено відповідних графів.")


```
